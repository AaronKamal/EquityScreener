Transformations

```python
def df_input(df):
    df['EXPENSES'] = df['Cost_of_Sales'] + df['Administration_and_Other_Expenses'] + df['Depreciation_and_Amortization_Cost']
    df['EARNING_BEFORE_INTEREST_AND_TAX'] = df['Revenue'] + df['EXPENSES']
    df['EARNING_BEFORE_TAX'] = df['EARNING_BEFORE_INTEREST_AND_TAX'] + df['Finance_cost']
    df['RETAINED_EARNINGS'] = df['EARNING_BEFORE_TAX'] + df['DIVIDEND_PAYOUT']
    df = df[['NUMBER_OF_YEARS_IN_BUSINESS', 'Revenue', 'EXPENSES',
            'EARNING_BEFORE_INTEREST_AND_TAX', 'Finance_cost', 'EARNING_BEFORE_TAX',
            'DIVIDEND_PAYOUT', 'RETAINED_EARNINGS', 'CURRENT_ASSETS', 'INVENTORY',
            'Non_Current_Assets', 'Short_term_Liabilities','Long_term_Liabilities']]
    return df
```

```python
input_df=df_input(data)
input_df.head(20)
```
