```python
#Credit score to operating revenue/turnover
import pandas as pd
import numpy as np


df = pd.read_csv("21_sectors_final.csv")
df['samenif']=np.nan
for i in range(1,(len(df))):
    if df.loc[i,'NIF_Code']==df.loc[i-1,'NIF_Code']:
        df.loc[i,'samenif']=1
    else:
        df.loc[i,'samenif']=0
        
df['samenif'].fillna('1.0', inplace=True)
df['changeinCS']=np.nan
df['changeinCS1']=np.nan
df['changeinrev']=np.nan
df['changeinrev1']=np.nan


for i in range(0,(len(df)-1)):
    if df.loc[i,'samenif']==1.0:
        df.loc[i-1,'changeinCS']=df.loc[i-1,'Credit_score']-df.loc[i,'Credit_score']
    else:
        df.loc[i,'changeinCS']=0
for i in range(0,(len(df)-1)):
    if df.loc[i,'changeinCS']!=np.nan:
        if df.loc[i,'changeinCS']>0:
            df.loc[i,'changeinCS1']=1
        elif df.loc[i,'changeinCS']<0:
            df.loc[i,'changeinCS1']=-1
        else:
            df.loc[i,'changeinCS1']=0
            
for i in range(0,(len(df))):
    if df.loc[i,'samenif']==1.0:
        df.loc[i-1,'changeinrev']=df.loc[i-1,'Operating_Revenue/Turnover']-df.loc[i,'Operating_Revenue/Turnover']
    else:
        df.loc[i,'changeinrev']=0
for i in range(0,(len(df))):
    if df.loc[i,'changeinrev']!=np.nan:
        if df.loc[i,'changeinrev']>0:
            df.loc[i,'changeinrev1']=1
        elif df.loc[i,'changeinrev']<0:
            df.loc[i,'changeinrev1']=-1
        else:
            df.loc[i,'changeinrev1']=0
df = df[~df['changeinrev'].isin([np.nan])]
def diff(row):
        if row['changeinrev1']==row['changeinCS1']:
            return "1"
        else:
            return "0"
df = df.assign(similarity=df.apply(diff, axis=1))
difference=df['similarity'].value_counts()       
dif0= difference['0']
if difference ['1'] <1:
    dif1=0
if difference ['1'] >=1:
    dif1= difference['1']
totalused= dif0+dif1
accuracy= dif1/totalused
cnae=df['CNAE_2009_Code']
blank= ' '
#df.to_csv('data.csv')
finaldata=[blank,cnae,totalused,dif1,dif0,accuracy]
print(finaldata)

```
