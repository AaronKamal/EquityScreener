```python
import pandas as pd
import numpy as np
df=pd.read_csv("Compiled Financial Data - Final2.csv")
def df_input(df):
    df['additional_costs']=df['Operating_Revenue/Turnover']-df['Cost_of_sales']-df['Depreciation']-df['Cost_of_Employees']
    df['EARNING_BEFORE_INTEREST_AND_TAX'] = df['Operating_Revenue/Turnover'] + df['additional_costs']
    df['EXPENSES'] = df['Cost_of_sales'] + df['Depreciation'] + df['Cost_of_Employees']+df['additional_costs']
    df['EARNING_BEFORE_TAX'] = df['EARNING_BEFORE_INTEREST_AND_TAX'] + df['Financial_P/L']
    df['RETAINED_EARNINGS'] = df['EARNING_BEFORE_TAX'] + df['Dividend']
    df = df[['Company_Name','Company_Age_(years)', 'Operating_Revenue/Turnover', 'EXPENSES',
            'EARNING_BEFORE_INTEREST_AND_TAX', 'Financial_P/L', 'EARNING_BEFORE_TAX',
            'Dividend', 'RETAINED_EARNINGS', 'Current_Assets', 'Stocks',
            'Fixed_Assets', 'Current_Liabilities','Non_Current_Liabilities','additional_costs','EBIT']]
    return df
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("Compiled Financial Data - Final2.csv")
df['Cost_of_sales']=df['Material_Costs']
#dividend
df['previous_dividend']=df['Other_Shareholders_Funds'].shift(-1)
df['Dividend'] = df['Other_Shareholders_Funds'] - df['P/L_for_Period'] - df['previous_dividend']
df['Dividend']=df['Dividend'].abs()
df['Dividend']=df['Dividend']*(-1)

df['ADDITIONAL_COSTS']=df['Operating_Revenue/Turnover']-df['Cost_of_sales']-df['Depreciation']-df['Cost_of_Employees']-df['EBIT']
#expenses
df['EXPENSES'] = df['Cost_of_sales'] + df['Depreciation'] + df['Cost_of_Employees']+df['ADDITIONAL_COSTS']
df['EXPENSES']=df['EXPENSES']*(-1)
df['EARNING_BEFORE_INTEREST_AND_TAX'] = df['Operating_Revenue/Turnover'] + df['EXPENSES']
df['EARNING_BEFORE_TAX'] = df['EARNING_BEFORE_INTEREST_AND_TAX'] + df['Financial_P/L']
df['RETAINED_EARNINGS'] = df['EARNING_BEFORE_TAX'] + df['Dividend']
df1 = df[['NIF_Code','Date','CNAE_2009_Code','Company_Age_(years)', 'Operating_Revenue/Turnover', 'EXPENSES',
            'EARNING_BEFORE_INTEREST_AND_TAX', 'Financial_P/L', 'EARNING_BEFORE_TAX',
            'Dividend', 'RETAINED_EARNINGS', 'Current_Assets', 'Stocks',
            'Fixed_Assets', 'Current_Liabilities','Non_Current_Liabilities','ADDITIONAL_COSTS']]
#df1.head(10)
df1.to_excel("Transformed_data.xlsx")
```
